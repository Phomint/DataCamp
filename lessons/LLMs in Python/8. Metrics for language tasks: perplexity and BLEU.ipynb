- ## BLEU
    - Measures translation quality against **human references**
    - Predictions: LLM's outputs
    - References: human references

```python
bleu = evaluate.load('bleu')

input_text = 'Lastest research findings in Antarctica show'
references = [["Lastest research findings in Antarctica show sifnificant ice loss due to climate change.", 'Lastest research findings in Antarctica show that the ice sheet is melting faster than previously thought.']]

generated_text = "Lastest research findings in Antarctica show that the ice sheet is melting faster than previously thought."
```

```python
results = bleu.compute(predictions = [generated_text], references = references)
print(results)
```


- 0-1 score: closer to 1 = higher similarity
