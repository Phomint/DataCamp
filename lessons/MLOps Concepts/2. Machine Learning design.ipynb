## Design phase

### Added value
- Estimate the expected value
- ML is experimental and uncertain
- Aids in resource allocation, prioritization, and setting expectations

### Business requirements
- **End user**
    -  **Speed** (we must consider the frequency of the predictions and how fast we need them.)
    -  **Accuracy** (we must also evaluate the model's accuracy and whether its results are explainable to non-experts)
    - **Transparency** (why the model makes its predictions, why it's wrong, and how we can improve the model)
- **Compliance and regulations** (depending on the problem we are trying to solve)
- **Budget** available
- **Team** size

### Key metrics
To see if the machine learning lifecycle progresses as expected. However, the roles involved in MLOps processes are multidisciplinary and thus also have their own way of tracking performance.
 - **Data scientist** looks at the accuracy of a model, how many times the algorithm is correct.
 - **Subject-matter** experts are interested in the model's impact on the business, for instance, how their work improves due to the use of machine learning.
 - **Business stakeholders** are more interested in the monetary value of the mode, in how many cases we actually generate revenue. This is often expressed in money or time.

To get the most out of machine learning, we must align the different metrics to make sure everyone is on the same page.
