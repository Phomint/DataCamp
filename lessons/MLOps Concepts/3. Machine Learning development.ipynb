## Feature engineering
... is the process of selecting, manipulating, and transforming raw data into features.
- A feature is a variable, such as the column in a table
- We can use raw data but also create our own

In this example, we can use `Number of orders` and `Total expenditure` to create a new feature `Average expenditure`

- The goal is to enhance model performance
- Tools and techniques help to process, select, and maintain features:
    - Feature selection
    - Feature store
    - Data version control

### Feature selection
- **Domain-specific knowledge** (This expertise helps ensure that the features used in the model are not only statistically significant, but also meaningful in the context of the problem)
- **Correlation** (Examine correlations to identify and eliminate redundant features.
- **Feature importance** (To understand which features significantly influnce the model's predictions.
- **Other methods**: univariate selection, Principal Component analysis (PCA), Recursive Feature Elimination (RFE)

## The feature store
... is a centralized repository for features, allowing data scientists to discover, define, and reuse features across projects.
- Essential in large teams where multiple projects need consistent and reusable features.

### The machine learning experiment
#### Why is experiment tracking important?
In each experiment, the following factors can be configured:
- Machine learning models
- Model hyperparameters
- Version of data
- Execution scripts
- Environment configurations

#### Using experiment tracking in the ML lifecycle is crucial?
It helps us compare results:
- Compare results
- Reproduce past experiments
- Collaborate with developers and stakeholders
- Report on results to stakeholders

### Example
Imagine we're building a model to classify images as either containing a dog or a cat.
In our first experiment, we might train a neural network with one hidden layer using 1000 images.

In the next, we add puppies and kittens to our dataset, bump it up to 2000 images, and use a model with two hidden layers.


A machine learning experiment will roughly go through the following steps:
- We begin by formulating a hypothesis based on our business objectives.
- We gather the necessary data, and then we set our initial hyperparameters.
- Define experiments, e.g., types of models, hyperparameters, datasets.
- We establish our experience tracking to log every detail meticulously.
- We proceed to train and evaluate the model, carefully storing information about the dataset and other relevant settings. Test the models on a hold-out test set.
- Once we've identified the best-performing model, we register it, along with all associated experiment details.
- Visualize and report back to team and stakeholders, and determine next steps
