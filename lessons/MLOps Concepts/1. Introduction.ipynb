## Machine Learning Operations
... is the set of practices to **design, deploy** and **maintain** machine learning in production **continuously, reliably,** and **efficiently**
- Focus on machine learning 'in production'

#### Machine Learning Life-cycle


## Why MLOps


- Improved collaboration
- Automation of deployment
- Monitoring of model performance

### Why the machine learning lifecycle?

- Structures the ML process
- Defines key players at each stage
- Toolkit for optimization

#### Design phase
- Context of the problem
- Added value
- Business requirements
- Key metrics
- Data processing

**(Engage stakeholders to evaluate the project's viability and meke informed decisions.)**

#### Development phase
- Develop machine learning model
- Experiment with data, algorithms, and hyperparameters
- Model ready for deployment

#### Deployment phase
- Integrate the machine learning model in business
- Deploying the model in production (ensuring it operates seamlessly within the larger system)
- We might build a microservice around our model (allowing for easy access and scalability)
- Monitoring the performance (Detect data drift, and receive alerts if the model's predictions begin to degrade)

### Business roles
- Business stakeholders
    - Budget decisions
    - Vision of a company
    - Involved throughout the lifecycle
- Subject-matter expert
    - Domain knowledge
    - Involved throughout the lifecycle (can assist the more technical roles with interpreting the data and results)

### Technical roles
- Data scientist
    - Data analysis
    - Model training and evaluation
- Data engineer
    - Collecting, storing, and processing data
    - Check and maintain data quality
- ML engineer
    - Versatile role
    - Specifically designed for complete machine learning lifecycle

### Extra roles
- Data analysts
- Developers
- Software engineers
- Backend engineers
(can all contribute significantly depending on the specific application of machine learning and the type of company)
