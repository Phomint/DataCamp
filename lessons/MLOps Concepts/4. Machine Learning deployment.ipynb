# Machine Learning Deployment


## Runtime Environments

### What is a Runtime Environment?
A runtime environment is the infrastructure and set of dependencies required to execute a machine learning model. It includes:

- **Operating System**: The base OS (Linux, Windows, macOS)
- **Language Runtime**: Python interpreter, R runtime, etc.
- **Libraries & Dependencies**: NumPy, Pandas, Scikit-learn, TensorFlow, PyTorch
- **System Libraries**: CUDA for GPU support, OpenBLAS for optimized computations
- **Environment Variables**: Configuration settings, API keys
- **Hardware**: CPU, GPU, memory specifications

## Environment Inconsistencies

### Common Problems

Machine learning models often face the "it works on my machine" problem. Key inconsistencies include:

1. **Dependency Conflicts**
   - Different library versions between development and production
   - Missing dependencies in production environment
   - Incompatible package versions

2. **System Differences**
   - Different operating systems
   - Different hardware configurations
   - Different Python versions

3. **Configuration Issues**
   - Missing environment variables
   - Different file paths
   - Network configurations

## Containers as a Solution

### What are Containers?

Containers are lightweight, standalone packages that include everything needed to run an application:
- Code
- Runtime
- System tools
- System libraries
- Settings

**Key Benefit**: "Package once, run anywhere"

## Docker: The Leading Container Platform

### Docker Advantages for ML Deployment

1. **Reproducibility**
   - Exact same environment across development, testing, and production
   - Version-controlled infrastructure

2. **Isolation**
   - Each model runs in its own container
   - No conflicts between different models' dependencies

3. **Portability**
   - Run on any system that supports Docker
   - Easy migration between cloud providers

4. **Scalability**
   - Easy to replicate containers for load balancing
   - Orchestration with Kubernetes for large-scale deployments

## Monolith vs. microservice architecture
Let's say we are running a webshop with a payment service, a service for the shopping cart, and a service for the inventory. We can run each service on the same computer, also known as a single instance **Monolith architecture**. In traditional software development, applications were often built in a monolithc architecture.
- One uniform application containing all services


A different solution is the **Microservices architecture** is, contrary to the monolithic architecture, a collection of smaller services that can be deployed independently.


If an application fails in a microservices architecture, only the separate service fails, while in a monolithic architecture, the whole application will fail.
A monolithic application can become complex since all services are intertwined and not independent. This also makes it harder to scale.
The usage of a microservices architecture depends on the application. If our application is very small, having a separate microservice for each even smaller part can become costly because each service requires compute power and has to be maintained independently

### Inferencing
It's common practice to deploy the machine learning model as a microservice. This allows us to use the machine learning model to make predictions based on new, unseen data. These processes are called inferencing.

... Inferencing is the process in which we send new input to the machine learning model and receive output from the model

### Application Programming Interface (API)
To provide communication between microservices, it is common practice to develop an API, an API is a set of pre-defined input and output combinations that allows different services to communicate.
