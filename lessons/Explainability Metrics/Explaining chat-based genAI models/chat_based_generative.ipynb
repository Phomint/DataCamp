### Self-consistency
Assesses model's confidence in generated answers
which allows us to access model's consistency for a specific outcomes

- this is specially used for classification tasks

### Creating self-consistency prompts

```python
prompt = """Classify the following review as positive or negative. You should reply with either 'positive' or 'negative', nothing else. Review: ' the customer service was great, but the product itself did not meet my expectations' """

responses = []
for i in range(5):
    sentiment = get_response(prompt)
    responses.append(sentiment.lower())

confidence = {
    'positive': responses.count('positive') / len(responses),
    'negative': responses.count('negative') / len(responses)
}

print(confidence)
```
