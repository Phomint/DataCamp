## Consistency
- Assesses stability of explanations when model trained on different subsets (training, validation or others)
- Low consistency -> no robust explanations


### Consine Similarity
- Near -1 opposite explanations highlithing potetial issues
- Near 0 no consistent explanations/patterns
- Near 1 consistent explanations/patterns
